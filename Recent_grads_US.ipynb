```python
import pandas as p
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
recent_grads = p.read_csv("recent-grads.csv")
recent_grads.iloc[:1]
recent_grads.info()
raw_data_count= recent_grads.shape[0] # Number of rows
print(raw_data_count)
```

```python
recent_grads.describe()
```

```python
## Drop the rows with missing data 

raw_data_count= recent_grads.shape[0] # Number of rows
print(raw_data_count)
# Drop rows containing missing values 
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print(cleaned_data_count)
```

```python
ax=recent_grads.plot(x="Sample_size",y="Median", kind="scatter")
ax.set_title("Sample size vs Median")
```

```python
ax=recent_grads.plot(x="Sample_size",y="Unemployment_rate", kind="scatter")
ax.set_title("Sample size vs Unemployment rate")
```

```python
x_t_values = list(recent_grads["Full_time"])

ax=recent_grads.plot(x="Full_time",y="Median", kind="scatter")
```

```python
ax=recent_grads.plot(x="ShareWomen",y="Median", kind="scatter")
ax.set_title(" ShareWomen vs Median ")
```

```python
ax=recent_grads.plot(x="Men",y="Median", kind="scatter")
ax.set_title(" Men vs Median ")
```

```python
ax=recent_grads.plot(x="Women",y="Median", kind="scatter")
ax.set_title(" Women vs Median ")
```

##### Generating graphs using Pandas (above examples)

- Generating the graph using Dataframe.plot() function.
- It can be customised using multiple parameters.
- It returns an Matplotlib.AxisSubplot object which can again be customised.

```python
var = ["Sample_size","Median","Employed","Full_time","ShareWomen","Unemployment_rate" ,"Men", "Women"]

fig = plt.figure(figsize=(10,60))
for r in range(0,8):
    ax = fig.add_subplot(12,1,r+1)
    ax = recent_grads[var[r]].plot(kind='hist', rot=40)
    ax.set_title(var[r])
```

##### Generating the plots using Series.plot() (above example)
- This method takes no argument, and generates the frequency dist graph
- Series.hist() could be used for more control

#### Generating graphs using Pandas.tools.plotting module (below example)
- Scatter matrix plot - combines scatter and matrix plot can be generated using scatter_matrix() function 
- scatter plot requires 2 vars and histogram requires 1 var (freq distribution)

```python
from pandas.tools.plotting import scatter_matrix

scatter_matrix(recent_grads[["Sample_size","Median"]], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[["Sample_size","Median","Unemployment_rate"]])
```
